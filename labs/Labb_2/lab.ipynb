#### b) Läs in 10 bilder slumpmässigt, plotta dem och extrahera deras labels och skriv ut i titeln.

```python
import os
import random


# Function to extract label from file path
def get_label(file_path):
    # Assuming the directory structure includes the label as the parent directory
    return os.path.basename(os.path.dirname(file_path))

# Load image file path
file_path = glob("../data/archive/train/train/*.jpg")

# Combine and shuffle
#all_files = cat_files + dog_files
random.shuffle(file_path)

# Select 10 random images
selected_files = random.sample(file_path, 10)

# Plot the images
plt.figure(figsize=(20, 10))

for i, file_path in enumerate(selected_files):
    # Read the image
    image = imread(file_path)
    # Extract the label
    label = get_label(file_path)
    # Create a subplot
    plt.subplot(2, 5, i + 1)
    plt.imshow(image)
    plt.title(label)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import shutil
import os


# Section 1: Move 1600 files to the experiment_small_data/train file
source_file = "../labs/Labb_2/Original_data/train/train"
destination_file = "../labs/Labb_2/experiment_small_data/train"
files_to_copy_train = 800
total_count_train = 0

for files in os.listdir(source_file):
    source_path = os.path.join(source_file, files)
    destination_path = os.path.join(destination_file, files)
    shutil.move(source_path, destination_path)
    total_count_train += 1
    if total_count_train >= files_to_copy_train:
        break

print(f"File '{source_file}' copied to '{destination_file}'")


# Section 2: Move 400 file to the experiment_small_data/val
source_file = "../labs/Labb_2/Original_data/train/train"
destination_file = "../labs/Labb_2/experiment_small_data/val"
files_to_copy_val = 400
total_count_val = 0

for file_val in os.listdir(source_file):
    if file_val.startswith("cat") or file_val.startswith("dog"):
        source_path = os.path.join(source_file, file_val)
        destination_path = os.path.join(destination_file, file_val)
        shutil.move(source_path, destination_path)
        total_count_val += 1
    if total_count_val >= files_to_copy_val:
        break

print(f"File '{source_file}' copied to '{destination_file}'")


# Section 3: Move 500 files to the experiment_small_data/test
source_file = "../labs/Labb_2/Original_data/train/train"
destination_file = "../labs/Labb_2/experiment_small_data/test"
files_per_category = 250
total_count_test = 0

for filename in os.listdir(source_file):
    if filename.startswith("cat") or filename.startswith("dog"):
        source_path = os.path.join(source_file, filename)
        destination_path = os.path.join(destination_file, filename)
        shutil.move(source_path, destination_path)
        total_count_test += 1
    if total_count_test >= 2 * files_per_category:
        break

print(f"File '{source_file}' copied to '{destination_file}'")
```

#### e) Läs in dataseten från experiment_small, experiment_tiny (BONUS) och plocka ut labelsvektorer, som ska vara one-hot encoded med 0 och 1.

```python
import os
import matplotlib.pyplot as plt
from pathlib import Path

# Define the datapath using Path and os
datapath = Path(os.path.join(Path().absolute().parent, "labs", "Labb_2", "experiment_small_data", "test"))

# List comprehension to read images from the datapath
images = [plt.imread(os.path.join(datapath, file)) for file in os.listdir(datapath) if file.endswith(".jpg")]

# Print the number of images
print(len(images))
```

```python
import numpy as np

# Assuming you have a function or variable named num_samples_code that provides the number of samples
num_samples = len(images)

# Number of classes
num_classes = 2  # Assuming 2 classes: "dog" and "cat"

# Generate random labels for the samples
labels = np.random.randint(0, num_classes, size=num_samples)

# One-hot encoding
label_vectors = np.eye(num_classes)[labels]

# Print the first few labels and their corresponding one-hot vectors for verification
print("Labels:")
print(labels[:10])  # Print first 10 labels
print("\nOne-hot Vectors:")
print(label_vectors[:10])  # Print one-hot vectors for first 10 labels
```

- plotta några bilder med deras respektive labels och kontrollera att det är korrekt.

```python
num_images_to_plot = 3

# Using a list comprehension with plt.subplots for plotting
fig, axs = plt.subplots(1, num_images_to_plot, figsize=(15, 5))

for ax, img, lbl, vec in zip(axs, images[:num_images_to_plot], labels[:num_images_to_plot], label_vectors[:num_images_to_plot]):
    ax.imshow(img)
    ax.set_title(f"Label: {lbl}, One-hot Vector: {vec}")
    ax.axis('off')

plt.show()
```

- skapa lämplig plot för att kontrollera att dataseten är balanserade

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert labels to a pandas Series for counting
labels_series = pd.Series(labels)

# Count the number of samples for each class
class_counts = labels_series.value_counts().sort_index()

# Plot a bar chart to visualize class distribution
class_counts.plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.title('Class Distribution')
plt.xticks(rotation=0)  # Ensure class labels are horizontal
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame for labels with class names
import pandas as pd
df = pd.DataFrame({
    'Labels': labels
})
df['Class'] = df['Labels'].map({0: 'Cat', 1: 'Dog'})

# Plot a histogram of the labels using seaborn
sns.histplot(data=df, x='Class', discrete=True, shrink=0.8, color='skyblue')
plt.title('Distribution of Labels')
plt.xlabel('Classes')
plt.ylabel('Number of Images')
plt.show()
```
