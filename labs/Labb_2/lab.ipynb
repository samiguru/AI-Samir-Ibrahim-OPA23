#### b) Läs in 10 bilder slumpmässigt, plotta dem och extrahera deras labels och skriv ut i titeln.

```python
import os
import random


# Function to extract label from file path
def get_label(file_path):
    # Assuming the directory structure includes the label as the parent directory
    return os.path.basename(os.path.dirname(file_path))

# Load image file path
file_path = glob("../data/archive/test/test/*.jpg")

# Combine and shuffle
#all_files = cat_files + dog_files
random.shuffle(file_path)

# Select 10 random images
selected_files = random.sample(file_path, 10)

# Plot the images
plt.figure(figsize=(20, 10))

for i, file_path in enumerate(selected_files):
    # Read the image
    image = imread(file_path)
    # Extract the label
    label = get_label(file_path)
    # Create a subplot
    plt.subplot(2, 5, i + 1)
    plt.imshow(image)
    plt.title(label)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
